```python
### 開始使用套件、定義聊天機器人
from chatterbot import ChatBot
chatbot = ChatBot("ChatBot_name")
```

```python
### 導入聊天機器人 - 文本訓練模組，並執行訓練
from chatterbot.trainers import ChatterBotCorpusTrainer

trainer = ChatterBotCorpusTrainer(chatbot)

trainer.train("/usr/local/lib/python3.7/dist-packages/chatterbot_corpus/data/tchinese")
```

```python
### 開始對話
ask = "你好嗎"
response = chatbot.get_response(ask)
print(response)
```
